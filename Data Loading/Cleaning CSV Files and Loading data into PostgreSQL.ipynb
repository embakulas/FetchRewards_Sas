```python
import pandas as pd
from datetime import datetime
```

```python
# Load CSV files into pandas dataframes
users_df = pd.read_csv(r'C:\Users\embak\OneDrive\Desktop\Fetch Assessment\Solution\Databricks output\users.csv')
receipts_df = pd.read_csv(r"C:\Users\embak\OneDrive\Desktop\Fetch Assessment\Solution\Databricks output\receipts.csv")
brands_df = pd.read_csv(r"C:\Users\embak\OneDrive\Desktop\Fetch Assessment\Solution\Databricks output\brands.csv")
brands1_df = pd.read_csv(r"C:\Users\embak\OneDrive\Desktop\Fetch Assessment\Solution\Databricks output\brands1.csv")
receiptItems_df = pd.read_csv(r"C:\Users\embak\OneDrive\Desktop\Fetch Assessment\Solution\Databricks output\receiptitems.csv")
```

```python
# Add default user to users_df
default_user_id = "default_user_id"
default_user = pd.DataFrame([{
    "user_id": default_user_id,
    "state": "N/A",
    "createdDate": pd.Timestamp.utcnow(),
    "lastLogin": pd.Timestamp.utcnow(),
    "role": "default",
    "active": True,
    "signUpSource": "default"
}])
```

```python
# Append the default user to users_df using concat
users_df = pd.concat([users_df, default_user], ignore_index=True)
```

```python
# Ensure user_id in receipts_df references users_df, if not set to default_user_id
receipts_df['user_id'] = receipts_df['user_id'].apply(lambda x: x if x in users_df['user_id'].values else default_user_id)
```

```python
# Merge brands_df and brands1_df
def merge_brands(brands_df, brands1_df):
    brands_df['brand_id'] = brands_df['brand_id'].astype('string')
    brands_df['barcode'] = brands_df['barcode'].astype('string')
    brands_df['brandCode'] = brands_df['brandCode'].astype('string')
    brands_df['category'] = brands_df['category'].astype('string')
    brands_df['categoryCode'] = brands_df['categoryCode'].astype('string')
    brands_df['name'] = brands_df['name'].astype('string')
    brands_df['topBrand'] = brands_df['topBrand'].astype(bool)
    brands_df['cpgId'] = brands_df['cpgId'].astype('string')
    brands_df['cpgRef'] = brands_df['cpgRef'].astype('string')

    brands1_df['brand_id'] = brands1_df['brand_id'].astype('string')
    brands1_df['barcode'] = brands1_df['barcode'].astype('string')
    brands1_df['brandCode'] = brands1_df['brandCode'].astype('string')
    brands1_df['category'] = brands1_df['category'].astype('string')
    brands1_df['categoryCode'] = brands1_df['categoryCode'].astype('string')
    brands1_df['name'] = brands1_df['name'].astype('string')
    brands1_df['topBrand'] = brands1_df['topBrand'].astype(bool)
    brands1_df['cpgId'] = brands1_df['cpgId'].astype('string')
    brands1_df['cpgRef'] = brands1_df['cpgRef'].astype('string')

    merged_brands_df = pd.concat([brands_df, brands1_df]).drop_duplicates().reset_index(drop=True)
    return merged_brands_df
```

```python
# Define function to transform users dataframe
def transform_users(users_df):
    users_df['user_id'] = users_df['user_id'].astype('string')
    users_df['state'] = users_df['state'].astype('string')
    users_df['createdDate'] = pd.to_datetime(users_df['createdDate'])
    users_df['lastLogin'] = pd.to_datetime(users_df['lastLogin'])
    users_df['role'] = users_df['role'].astype('string')
    users_df['active'] = users_df['active'].astype(bool)
    users_df['signUpSource'] = users_df['signUpSource'].astype('string')
    return users_df
```

```python
# Define function to transform receipts dataframe
def transform_receipts(receipts_df):
    receipts_df['receipt_id'] = receipts_df['receipt_id'].astype('string')
    receipts_df['user_id'] = receipts_df['user_id'].astype('string')
    receipts_df['bonusPointsEarned'] = receipts_df['bonusPointsEarned'].fillna(0).astype(int)
    receipts_df['bonusPointsEarnedReason'] = receipts_df['bonusPointsEarnedReason'].astype('string')
    receipts_df['createDate'] = pd.to_datetime(receipts_df['createDate'])
    receipts_df['dateScanned'] = pd.to_datetime(receipts_df['dateScanned'])
    receipts_df['finishedDate'] = pd.to_datetime(receipts_df['finishedDate'])
    receipts_df['modifyDate'] = pd.to_datetime(receipts_df['modifyDate'])
    receipts_df['pointsAwardedDate'] = pd.to_datetime(receipts_df['pointsAwardedDate'])
    receipts_df['pointsEarned'] = receipts_df['pointsEarned'].fillna(0.0).astype(float)
    receipts_df['purchaseDate'] = pd.to_datetime(receipts_df['purchaseDate'])
    receipts_df['purchasedItemCount'] = receipts_df['purchasedItemCount'].fillna(0).astype(int)
    receipts_df['rewardsReceiptStatus'] = receipts_df['rewardsReceiptStatus'].astype('string')
    receipts_df['totalSpent'] = receipts_df['totalSpent'].fillna(0.0).astype(float)
    return receipts_df
```

```python
# Define function to transform receipt items dataframe
def transform_receipt_items(receiptItems_df):
    receiptItems_df['receipt_id'] = receiptItems_df['receipt_id'].astype('string')
    receiptItems_df['barcode'] = receiptItems_df['barcode'].astype('string')
    receiptItems_df['description'] = receiptItems_df['description'].astype('string')
    receiptItems_df['finalPrice'] = receiptItems_df['finalPrice'].fillna(0.0).astype(float)
    receiptItems_df['itemPrice'] = receiptItems_df['itemPrice'].fillna(0.0).astype(float)
    receiptItems_df['needsFetchReview'] = receiptItems_df['needsFetchReview'].astype(bool)
    receiptItems_df['partnerItemId'] = receiptItems_df['partnerItemId'].fillna('0').astype('string')
    receiptItems_df['preventTargetGapPoints'] = receiptItems_df['preventTargetGapPoints'].astype(bool)
    receiptItems_df['quantityPurchased'] = receiptItems_df['quantityPurchased'].fillna(0).astype(int)
    receiptItems_df['userFlaggedBarcode'] = receiptItems_df['userFlaggedBarcode'].fillna('').astype('string')
    receiptItems_df['userFlaggedNewItem'] = receiptItems_df['userFlaggedNewItem'].astype(bool)
    receiptItems_df['userFlaggedPrice'] = receiptItems_df['userFlaggedPrice'].fillna(0.0).astype(float)
    receiptItems_df['userFlaggedQuantity'] = receiptItems_df['userFlaggedQuantity'].fillna(0).astype(int)
    receiptItems_df['needsFetchReviewReason'] = receiptItems_df['needsFetchReviewReason'].astype('string')
    receiptItems_df['pointsNotAwardedReason'] = receiptItems_df['pointsNotAwardedReason'].astype('string')
    receiptItems_df['pointsPayerId'] = receiptItems_df['pointsPayerId'].fillna('-1').astype('string')
    receiptItems_df['rewardsGroup'] = receiptItems_df['rewardsGroup'].astype('string')
    receiptItems_df['rewardsProductPartnerId'] = receiptItems_df['rewardsProductPartnerId'].astype('string')
    receiptItems_df['userFlaggedDescription'] = receiptItems_df['userFlaggedDescription'].astype('string')
    return receiptItems_df
```

```python
# Call functions to transform each dataframe
merged_brands_df = merge_brands(brands_df, brands1_df)
users_df = transform_users(users_df)
receipts_df = transform_receipts(receipts_df)
receiptItems_df = transform_receipt_items(receiptItems_df)
```

```python
merged_brands_df = merged_brands_df.drop_duplicates()
users_df = users_df.drop_duplicates()
receipts_df = receipts_df.drop_duplicates()
receiptItems_df = receiptItems_df.drop_duplicates()
```
